```python
from sklearn.datasets import fetch_openml
import matplotlib.pyplot as plt
import numpy as np
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score, f1_score, recall_score, precision_score, RocCurveDisplay
from sklearn import metrics
```

```python
X,y=fetch_openml(data_id=337, return_X_y=True)

X=X.to_numpy()
y=y.to_numpy()
y=y.astype(int)

clf=MLPClassifier(hidden_layer_sizes=(45,45,45), max_iter=500).fit(X,y)
y_hat=clf.predict(X)

acc=accuracy_score(y,y_hat)
f1=f1_score(y,y_hat)
recall=recall_score(y,y_hat)
prec=precision_score(y,y_hat)
print("The accuracy is %s, the f1 score is %s, the recall score is %s and the precision score is %s" % (acc,f1,recall,prec))
```

```python
fpr, tpr, thresholds = metrics.roc_curve(y, y_hat)
roc_auc = metrics.auc(fpr, tpr)
display=metrics.RocCurveDisplay(fpr=fpr, tpr=tpr, roc_auc=roc_auc, estimator_name='ROC-curve')
display.plot()
plt.plot([0, 1], [0, 1], color="navy", linestyle="--")
plt.show()
```
